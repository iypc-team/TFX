```python
from __future__ import absolute_import, division
from IPython.display import clear_output
from BashColors import C
from time import perf_counter, sleep
import concurrent.futures, sys, threading
from concurrent.futures import ThreadPoolExecutor
mods=sys.modules.keys()

def getTensorflow():
    %pip install tensorflow
    print()
    sleep(1)
    return f'installed: {C.BIBlue}tensorflow{C.ColorOff}'

def getTFHub():
    %pip install --upgrade tensorflow-hub
    print()
    sleep(1)
    return f'installed: {C.BIBlue}tensorflow-hub{C.ColorOff}'
    
def getTFLite():
    %pip install tflite-model-maker
    print()
    sleep(1)
    return f'installed: {C.BIBlue}tflite-model-maker{C.ColorOff}'
    
def getResults():
    %pip install --upgrade results
    print()
    sleep(1)
    return f'installed: {C.BIBlue}results{C.ColorOff}'

def getNumpy():
    %pip install numpy
    print()
    sleep(1)
    return f'installed: {C.BIBlue}numpy{C.ColorOff}'
    
def getCV2():
    %pip install opencv-python-headless # works.
    print()
    sleep(1)
    return f'installed: {C.BIBlue}opencv-python-headless{C.ColorOff}'
    # %pip install opencv-contrib-python-headless

def getMatplotlib():
    %pip install matplotlib
    print()
    sleep(1)
    return f'installed: {C.BIBlue}matplotlib{C.ColorOff}'
    
def getSciPi():
    %pip install SciPy
    print()
    sleep(1)
    return f'installed: {C.BIBlue}SciPy{C.ColorOff}'

startTime = perf_counter()

with ThreadPoolExecutor() as executor:
    f1 = executor.submit(getTensorflow)
    f2 = executor.submit(getTFHub)
    f3 = executor.submit(getTFLite)
    f4 = executor.submit(getNumpy)
    f5 = executor.submit(getCV2)
    f6 = executor.submit(getMatplotlib)
    f7 = executor.submit(getSciPi)
    f8 = executor.submit(getResults)

finishTime = perf_counter()
clear_output(wait=1)
    
print(f1.result())
print(f2.result())
print(f3.result())
print(f4.result())
print(f5.result())
print(f6.result())
print(f7.result())
print(f8.result())
totalTime = finishTime - startTime
print(f'total time: {round(totalTime, 1)} second(s)')
```
